```python
# Handling Missing Values
  # False means no missing values
# df.isnull()
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
# Fill missing values with the mean of the column
df['new_sales_column'] = df['Sales'].fillna(df['Sales'].mean())

df
```

```python
df.dtypes
```

```python
# Renaming Columns
df = df.rename(columns={'Date': 'Sales Date'})
df.head()
```

```python
# Change Data Types
df['Value _new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
# Data Aggregating and Grouping
grouped_mean = df.groupby('Product')['Value'].mean()

print(grouped_mean)
```

```python
df.groupby(['Product', 'Region'])['Value'].mean()
```

```python
# Aggregate multiple functions
df.groupby('Sales')['Value'].agg(['mean', 'sum', 'count'])
```
