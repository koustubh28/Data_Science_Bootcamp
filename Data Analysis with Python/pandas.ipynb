Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary structures: Series and DataFrame.

A Series is a 1-d array like object, while a DataFrame is a 2-d, size-mutable, and potentially heterogeneous tabular data structure with labelled axis (rows and columns).

```python
import pandas as pd
```

```python
### Series - 1D Array Object
data = [1, 2, 3, 4, 5]

series = pd.Series(data) 
print(series)
```

```python
# Create a Series from a Dictionary
data = {'a': 1, 'b': 2, 'c': 3}
series_dict = pd.Series(data)
print(series_dict)
```

```python
ndata = {
  'Name': ['Krish', 'John', 'Jack'],
  'Age': [25, 30, 45],
  'City': ['Banglore', 'New York', 'Florida']
}

df = pd.DataFrame(ndata)
print(df)
print(type(df))
```

```python
print(df.at[1, 'Age'])
```
