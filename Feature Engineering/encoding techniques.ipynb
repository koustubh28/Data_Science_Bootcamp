Nominal/One Hot Encoding Technique

Basically we try to convert a categorical feature to numerical feature. That's called encoding.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create a dataframe
df = pd.DataFrame({
  'color': ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
# Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
# Perform fit and transform
encoder.fit_transform(df[['color']])
```

```python
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

encoder_df
```

```python
pd.concat([df, encoder_df], axis=1)
```

```python
df.head()
```

Ordinal Encoding:

It is used to encode categorical data that have an intrinsic order or ranking.

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
# Create a small dataframe
df = pd.DataFrame({
  'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})
```

```python
df
```

```python
# Create an instance of OrdinalEncoder and then fit_transform
encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoder.fit_transform(df[['size']])
```

Target Guided Ordinal Encoding

Here, we replace each category in the categorical variable with a numerical value based on the mean or median of the target variable for that category. 

```python
# Create a sample dataframe with a categorical variable and a target variable
df = pd.DataFrame({
  'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
  'price': [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
mean_price = df.groupby('city')['price'].mean()
```

```python
mean_price
```

```python
# Convert to dictionary
mean_price.to_dict()
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df[['price', 'city_encoded']]
```
