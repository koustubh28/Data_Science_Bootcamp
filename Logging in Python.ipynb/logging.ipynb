Logging is a crucial aspect of any application, providing a way to track events, errors, and operational information. Python's built-in logging module offers a flexible framework for emitting log messages from python programs.

We learn how to configure logging, log levels, and best practices for using logging in python applications.

```python
import logging 

# Configure the basic logging settings
logging.basicConfig(level=logging.DEBUG)

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

Log Levels

Python's logging modules has several log levels indicating the severity of events. 
The default levels are:

DEBUG: Detailed information, typically of interest only when diagnosing problems.

INFO: Confirmation things are working as expected.

WARNING: An indication that something unexpected happened or indicative of some problems in the near future.

ERROR: Due to a more serious problem, the software is not able to function properly.

CRITICAL: A very serious error, indicating that the program itself may unable to run.

```python
import logging

# Configuring logging
logging.basicConfig(
  filename='app.log',
  filemode='w',
  level=logging.DEBUG,
  format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
  datefmt='%y-%m-%d %H:%M:%S'
)

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
